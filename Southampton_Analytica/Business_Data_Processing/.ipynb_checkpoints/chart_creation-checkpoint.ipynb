```python
import pandas as pd
```

```python
data = pd.read_csv('pers_rating_biz_distrib.csv')
```

```python
data.columns
```

```python
data['I_tot'] = data['I1S'] + data['I2S'] + data['I3S'] + data['I4S'] + data['I5S']
data['E_tot'] = data['E1S'] + data['E2S'] + data['E3S'] + data['E4S'] + data['E5S']
data['S_tot'] = data['S1S'] + data['S2S'] + data['S3S'] + data['S4S'] + data['S5S']
data['N_tot'] = data['N1S'] + data['N2S'] + data['N3S'] + data['N4S'] + data['N5S']
data['F_tot'] = data['F1S'] + data['F2S'] + data['F3S'] + data['F4S'] + data['F5S']
data['T_tot'] = data['T1S'] + data['T2S'] + data['T3S'] + data['T4S'] + data['T5S']
data['J_tot'] = data['J1S'] + data['J2S'] + data['J3S'] + data['J4S'] + data['J5S']
data['P_tot'] = data['P1S'] + data['P2S'] + data['P3S'] + data['P4S'] + data['P5S']
```

```python
b_1 = data.iloc[0]
b_2 = data.iloc[1]
```

```python
data['I_1per'] = data['I1S'] / data['I_tot']
data['I_2per'] = data['I2S'] / data['I_tot']
data['I_3per'] = data['I3S'] / data['I_tot']
data['I_4per'] = data['I4S'] / data['I_tot']
data['I_5per'] = data['I5S'] / data['I_tot']


data['E_1per'] = data['E1S'] / data['E_tot']
data['E_2per'] = data['E2S'] / data['E_tot']
data['E_3per'] = data['E3S'] / data['E_tot']
data['E_4per'] = data['E4S'] / data['E_tot']
data['E_5per'] = data['E5S'] / data['E_tot']


data['N_1per'] = data['N1S'] / data['N_tot']
data['N_2per'] = data['N2S'] / data['N_tot']
data['N_3per'] = data['N3S'] / data['N_tot']
data['N_4per'] = data['N4S'] / data['N_tot']
data['N_5per'] = data['N5S'] / data['N_tot']


data['S_1per'] = data['S1S'] / data['S_tot']
data['S_2per'] = data['S2S'] / data['S_tot']
data['S_3per'] = data['S3S'] / data['S_tot']
data['S_4per'] = data['S4S'] / data['S_tot']
data['S_5per'] = data['S5S'] / data['S_tot']


data['F_1per'] = data['F1S'] / data['F_tot']
data['F_2per'] = data['F2S'] / data['F_tot']
data['F_3per'] = data['F3S'] / data['F_tot']
data['F_4per'] = data['F4S'] / data['F_tot']
data['F_5per'] = data['F5S'] / data['F_tot']


data['T_1per'] = data['T1S'] / data['T_tot']
data['T_2per'] = data['T2S'] / data['T_tot']
data['T_3per'] = data['T3S'] / data['T_tot']
data['T_4per'] = data['T4S'] / data['T_tot']
data['T_5per'] = data['T5S'] / data['T_tot']


data['P_1per'] = data['P1S'] / data['P_tot']
data['P_2per'] = data['P2S'] / data['P_tot']
data['P_3per'] = data['P3S'] / data['P_tot']
data['P_4per'] = data['P4S'] / data['P_tot']
data['P_5per'] = data['P5S'] / data['P_tot']


data['J_1per'] = data['J1S'] / data['J_tot']
data['J_2per'] = data['JS'] / data['J_tot']
data['J_3per'] = data['J3S'] / data['J_tot']
data['J_4per'] = data['J4S'] / data['J_tot']
data['J_5per'] = data['J5S'] / data['J_tot']
```

```python
b_1 = data.iloc[275]
b_2 = data.iloc[35] #275
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

b_1
```

```python
def get_val_set(b):
    e = [b['E_1per'], b['E_2per'], b['E_3per'], b['E_4per'], b['E_5per']]
    i = [b['I_1per'], b['I_2per'], b['I_3per'], b['I_4per'], b['I_5per']]
    n = [b['N_1per'], b['N_2per'], b['N_3per'], b['N_4per'], b['N_5per']]
    s = [b['S_1per'], b['S_2per'], b['S_3per'], b['S_4per'], b['S_5per']]
    return e,i,n,s
    
b1e, b1i, b1n, b1s = get_val_set(b_1)
b2e, b2i, b2n, b2s = get_val_set(b_2)
```

```python
f, ax = plt.subplots(2,2, figsize=(20, 20))
plt.subplots_adjust(hspace = 0.2, wspace = 0.1)
xi = [1,2,3,4,5]
a = ax[0,0].plot(b1e, label = 'b1', linewidth=4)
ax[0,0].set_xticks(xi)
ax[0,0].set_xlabel('Rating', fontsize=15)
ax[0,0].set_ylabel('Probability', fontsize=15)
ax[0,0].set_ylim([0, 0.4])
ax[0,1].set_xlabel('Rating', fontsize=15)
ax[0,1].set_ylabel('Probability', fontsize=15)
ax[0,1].set_ylim([0, 0.4])
ax[1,0].set_xlabel('Rating', fontsize=15)
ax[1,0].set_ylabel('Probability', fontsize=15)
ax[1,0].set_ylim([0, 0.4])
ax[1,1].set_xlabel('Rating', fontsize=15)
ax[1,1].set_ylabel('Probability', fontsize=15)


ax[1,1].set_ylim([0, 0.4])
b = ax[0,0].plot(b2e, label = 'b2', linewidth=4)
ax[0,0].set_title('Extrovert', fontsize=25)
c = ax[0,1].plot(b1i, label = 'b1', linewidth=4)
d = ax[0,1].plot(b2i, label = 'b2', linewidth=4)
ax[0,1].set_title('Introvert', fontsize=25)
e = ax[1,0].plot(b1n, label = 'b1', linewidth=4)
f = ax[1,0].plot(b2n, label = 'b2', linewidth=4)
ax[1,0].set_title('Intuition', fontsize=25)
g = ax[1,1].plot(b1s, label = 'b1', linewidth=4)
h = ax[1,1].plot(b2s, label = 'b2', linewidth=4)
ax[1,1].set_title('Sensing', fontsize=25)
ax[1,1].legend()
plt.savefig('ratings.pdf')
```
